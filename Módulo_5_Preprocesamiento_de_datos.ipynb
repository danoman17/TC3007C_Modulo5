```python
# importamos archivo csv de nombres de las firmas que contienen las firmas, con informacion relevante de cada una de estas.
data = pd.read_csv('us2022q2a.csv')
df1 = pd.read_csv('us2022q2a.csv',names = data.columns)
df1 # se imprime el dataset, para ver si es el correcto
```

```python
# En esta seccion se lee el otro archivo csv, donde podremos obtener, el nombre de las formas, el ticker y el sector al que 
# pertenece la firma correspondiente
data = pd.read_csv('usfirms2022.csv')
df2 = pd.read_csv('usfirms2022.csv', names = data.columns)
df2 = df2[["Ticker","Name", "Sector NAICS\nlevel 1"]] #definimos las colunas que se usaran para armar el nuevo dataframe
df2 # de igual manera imprimimos el dataset
```

```python
# a traves de la función merge, juntamos estos dos datasets separados anteriormente
df = pd.merge(df1, df2, left_on='firm', right_on="Ticker")

# al tener mergeado ambos datasets, se procede a dropear filas que contengan valores nulos, y la columna de Ticker, ya que esta solo fue usada 
# .. para poder usarla de indice para mergear los datasets

#cabe mencionar que solo se realiza a las columnas más importantes del dataset
df = df.drop(columns=['Ticker'])
df = df.drop(df[df.firm.isnull()].index)
df = df.drop(df[df.revenue.isnull()].index)
df = df.drop(df[df.cogs.isnull()].index)
df = df.drop(df[df.adjprice.isnull()].index)
df #imprimimos el data set para visualizar como quedó
```

```python
#calculamos las nuevas columnas, a traves de la funcion assign y a traves de formulas proporcionadas en clase:

# Market Value = (originalprice * sharesoutstanding)
df = df.assign(market_value=lambda x: x.originalprice.astype(np.float) * x.sharesoutstanding.astype(np.float))

# Book Value of the firm = (totalassets-totalliabilities)
df = df.assign(book_value=lambda x: x.totalassets.astype(np.float) - x.totalliabilities.astype(np.float))

# Print dfF y para observar si las nuevas columnas de BookValue y MarkertValue fueron creados de manera correcta
df
```

```python
#Agregmamos book to market ratio al dataset en donde estamos trabajando
#BookToMarketRatio = bookvalue / marketvalue
df = df.assign(BookToMarketRatio=lambda x: x.book_value.astype(np.float) / x.market_value.astype(np.float))
df 
```

```python
#ShortFinancialLeverage = shortdebt / totalassets

df = df.assign(ShortFinancialLeverage=lambda x: x.shortdebt.astype(np.float) / x.totalassets.astype(np.float) )
df
```

```python
#LongFinancialLeverage = longdebt / totalassests

df = df.assign(LongFinancialLeverage=lambda x: x.longdebt.astype(np.float) / x.totalassets.astype(np.float) )
df
```

```python
df.columns
```

```python
#EPS = netincome / #SharesOutstanding
#netincome = revenue - cogs - sgae - otheropexp - incometax - finexp + extraincome

df = df.assign(netincome=lambda x: x.revenue.astype(np.float) - x.cogs.astype(np.float) - x.sgae.astype(np.float) - x.otheropexp.astype(np.float) - x.incometax.astype(np.float) - x.finexp.astype(np.float) + x.extraincome.astype(np.float))

df = df.assign(EPS=lambda x: x.netincome.astype(np.float) / x.sharesoutstanding.astype(np.float) )

df['EPS']
```

```python
#creamos nueva matriz para poder calcular la matriz de varianza covarianza.

matrixOriginal = df_commercial[['EPSP','BookToMarketRatio','ShortFinancialLeverage','LongFinancialLeverage']]
matOriginal = matrixOriginal.dropna()
matOriginal
```

```python
#se realiza el calculo de la matriz de varianza covarianza con algebra matricial


X = matrixOriginal


n = X.count()['EPSP']
X_trans = X.T

shape = X.shape

M_x = X_trans.dot(np.ones((shape[0],1)))


varCov = (1/(n-1)) * (X_trans.dot(X) - (1/n) * (M_x.dot(M_x.T)))

varCov
```

```python
# calculamos la matriz con funciones de numpy para poder comprovar el resultado de la variable
X.cov()
```

```python
#calculamos la correlacion del las variables con algebra matricial

df_matrix_std = np.matrix(X.std())
corr_matrix = varCov / (df_matrix_std.T * df_matrix_std)
corr_matrix
```

1.2 Corre pruebas estadísticas para detectar outliers y leverage points. Tiene que utilizar álgebra matricial para las pruebas y explicar claramente cómo funcionan las pruebas. Puede utilizar funciones de Python para corroborar resultados.

**Calculo de Leverages**

```python
df_leverage_inicial = matOriginal[['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']]
df_leverage_inicial = df_leverage_inicial.dropna().reset_index(drop="True")
x = np.matrix(df_leverage_inicial)

x_ones = np.ones((df_leverage_inicial.count()['EPSP'], 1))


x = np.c_[x_ones, x]
h = x * np.linalg.inv(x.T * x) * x.T

df_leverage = pd.DataFrame(data = {'Leverage': np.diag(h)})
df_leverage = pd.concat([df_leverage_inicial, df_leverage], axis=1, join = "inner")

thrh = 3 * (len(['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']) + 1) / n
df_leverage[df_leverage['Leverage'] > thrh]
```

```python
import statsmodels.api as sm
```

```python
df_commercial = df_commercial.dropna()
df_commercial['R'] = np.log(df_commercial['adjprice'].astype(np.float)) - np.log(df_commercial.groupby(['firm'])['adjprice'].shift(4).astype(np.float))
df_commercial['R'] = df_commercial['R'].fillna(0)
```

```python


var_dep = ['R']
var_ind = ['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']


modelo = sm.OLS(df_commercial[var_dep], sm.add_constant(df_commercial[var_ind])).fit()

inf = modelo.get_influence()

lev = inf.hat_matrix_diag

sm_leverage = pd.DataFrame(data={'Leverage':lev})

sm_leverage

sm_leverage = pd.concat([df_commercial[var_ind].reset_index(drop="True"), sm_leverage], axis = 1, join ="inner")

sm_leverage[sm_leverage['Leverage'] > thrh]
```
