```python
#creamos una nueva matriz para poder calcular la matriz de varianza covarianza y limpiamos los datos de valores NaN.

matrixOriginal = df_commercial[['EPSP','BookToMarketRatio','ShortFinancialLeverage','LongFinancialLeverage']]
matOriginal = matrixOriginal.dropna()
matOriginal
```

```python
#se realiza el cálculo de la matriz de varianza-covarianza con algebra matricial.
# usaremos una formula para poder calcular la matriz, por lo que se caluclaran las variables a usar.

X = matrixOriginal
n = X.count()['EPSP']
X_trans = X.T

shape = X.shape

M_x = X_trans.dot(np.ones((shape[0],1)))


varCov = (1/(n-1)) * (X_trans.dot(X) - (1/n) * (M_x.dot(M_x.T)))

varCov
```

```python
# calculamos la matriz con funciones de numpy para poder comprovar el resultado de la variable
X.cov()
```

- ¿Que es la Varianza?
Podemos decir que la varianza es una medida de dispersion que se usa en estadistica y otras ciencias para poder representar la variabilidad de un conjunto de datos respecto de la media de estos mismos. Es por esto mismo que se calcula como la suma de los residuos elevados al cuadrado entre el total de observaciones ( el número de datos en el dataset o tabla).

- ¿Qué es la covarianza?
Podemos decir que es un valor en el cual se refleja en que cuantía o varian dos variables aleatorias varían de forma conjunta de acuerdo a sus medias, basicamente nos permite saber como se comporta una variable respecto al comportamiento de otra.

```python
#calculamos la correlacion del las variables con algebra matricial

matrix_std = np.matrix(X.std())
corr_matrix = varCov / (matrix_std.T * matrix_std)
corr_matrix
```

```python
#calculamos la correlación.
X.corr()
```

- **¿Qué es la correlación?**

Es una medida que nos dice el tipo de asociación entre dos variables numericas, espcificamente evalúa si la tendencia de los datos, es decir, si la relacion tiene un crecimiento creciante o decreciente.

**Interpretando la matriz de correlacion**

La manera de leer la matriz de correlaciones es comprender que la diagonal representa aquellos valores que son exactos (1), esto tiene sentido, pues como se observa, cada variable de columna y fila son correspondientes, es decir, son las mismas variables relacionandoce entre sí, esto logicamente dara un valor perfecto de correlación entre esas variables. Por otro lado, los otros valores arriba de los valores de la diagonal representan aquellos valores que se relacionan con otras variables distintas.

De manera resumida y rápida podemos decir que al tener un valor positivo > 0 quiere decir que estas tienen una relacion positiva entre una y otra de manera creciente, entre más cercano sea el valor a 1, podemos decir que estas variables forman una relacion fuerte, y entre mas cercano sea el valor a 0, podemos decir que estas tienen una relacion baja. Del mismo modo, pasa con valores negativos, solo que en vez de tener un valor positivo creciente, estas tendran un valor negativo decreciente. 


Con esto podemos interpretar nuestra matriz de correlaciones:

- Podemos observar que la variable de EPSP no tiene una correlacion alta con ninguna de las otras variables, ademas de que todas estos valores son negativos, lo que quiere decir que tiene un valor de pendiente bajo decreciente.

#**Cálculo de Leverages y Outliers**

Para la primer prueba que haremos para detectar o identidicar leverages será usar este metodo usando algebra matricial, con lo que podremos tomar la diagonal de la matriz obtenida, para asi poder calcular cuales son aquellos puntos más extremos en el eje x.

Basicamente podemos decir que todo parte de la fórmula para poder calcular las predicciones, es decir y hat, la cual podemos despejar de tal manera que nos quede una formula con los cálculos de los coeficientes estimaos representados con algebra matricial de manera que nos queda: 

y_hat= H*y

```python
# seleccionamos las variables con las que vamos a trabajar para calcular los laverage points
df_leverage_inicial = matOriginal[['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']]
df_leverage_inicial = df_leverage_inicial.dropna().reset_index(drop="True")
x = np.matrix(df_leverage_inicial)

# realizamos una matriz de ones
x_ones = np.ones((df_leverage_inicial.count()['EPSP'], 1))

#concatenamos la matriz con la matriz de ones
x = np.c_[x_ones, x]

# aplicamos la formula para obtener H, es decir, los leverage de cada punto de cada variable
h = x * np.linalg.inv(x.T * x) * x.T

#Finalmente creamos un dataframe con la diagonal de la matriz caculada el leverage h_ii.
df_leverage = pd.DataFrame(data = {'Leverage': np.diag(h)})

#lo concatenamos con la matriz calculada al inicio
df_leverage = pd.concat([df_leverage_inicial, df_leverage], axis=1, join = "inner")


#identificamos cuales valores son los mas extremos y los mostramos
thrh = 3 * (len(['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']) + 1) / n
df_leverage[df_leverage['Leverage'] > thrh]
```

Podemos observar en la grafica anteior que realmente se obtienen pocos valores extremos haciendo uso de los leverages, con un total de 13 valores extremos en el valor de x.

Aquí se usará el metodo de los residuales, con lo cual lo que queremos hacer es detectar aquellos valores que tienen un valor atípico en el resultado de los valores observados y que pueden afectar al momento de la realizacion de algun modelo. Ahora bien, para poder representar estos valores en el eje y, debemos realizar una regresion lineal, para posteriormente poder calcular los outliers con el método mencionado anteriormente.

```python
import statsmodels.api as sm
```

```python
#Dividimos las variables en independientes y dependientes para la realizacion del modelo de predicción.
var_dep = ['R']
var_ind = ['EPSP', 'BookToMarketRatio', 'ShortFinancialLeverage', 'LongFinancialLeverage']

#realizamos un modelo de prediccion lineal, con las variables dependientes e independientes
modelo = sm.OLS(df_commercial[var_dep], sm.add_constant(df_commercial[var_ind])).fit()

#obtenemos la infuencia de este modelo
influence = modelo.get_influence()

# calculamos nuestros valores de leverage y tomamos los resultados de la diagonal, es decir los h_ii.
lev = influence.hat_matrix_diag

#creamos el dataframe de leverage y lo concatenamos a la matris de df_commercial
smodel_leverage = pd.DataFrame(data={'Leverage':lev})

smodel_leverage = pd.concat([df_commercial[var_ind].reset_index(drop="True"), smodel_leverage], axis = 1, join ="inner")

smodel_leverage[smodel_leverage['Leverage'] > thrh] # solo mostramos aquellos que son valores extremadamente altos.
```

En la tabla de arriba podemos observar pocos valores extremos del modelo, en total 13, estos valores, como se menciona anteriormente, se debe a que el valor rebasa 3 veces la media de los leverage.

```python
print(modelo.summary())
```

```python
#Realizamos una prediccion con el modelo que acabamos de declara en la seccion de arriba 
prediccions = modelo.predict(sm.add_constant(df_commercial[var_ind]))

#calulamos los errores de los valores reales y observados
errores = np.matrix(df_commercial[var_dep]).T - np.matrix(prediccions)

#Obtenemos el MSE para completar la formula de los residuales
mse = np.square(errores).sum() / (df_commercial[var_ind].count()['EPSP'] - (len(var_ind) + 1))

#obtenemos la raiz del mse para completar la formula 
squared_error = np.sqrt(mse * (1 - lev))

# aplicamos la formula de los residuales standar.
std_residuals = pd.DataFrame((errores / squared_error).T)
std_residuals.columns = ['std_residuals'] #creamos una columna en donde se colocaran los std_residuals
std_residuals = pd.concat([df_commercial['R'].reset_index(drop="True"), std_residuals], axis = 1)
std_residuals = std_residuals.dropna()
std_residuals[std_residuals['std_residuals'] > 3]
```

Observamos que solo tenemos un valor atipico en el eje y, teniendo un valor de casi 3.7, en esa observación, lo que nos dice que la prediccion del modelo solo tiene 1 outlier
